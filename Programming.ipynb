Knowing these may be of no use to your job, but this is what is used to weed out non-hardworking people for an interview. Follow below steps to solve a problem.

1. When an interviewer says the problem, write all the key details at the top, like sorted array, non-decreasing etc... This will show how organized you are.
2. Make sure to to double check with the interviewer about the inputs and the outputs
3. What is the most important value of the problem? How much time, space and memory you have?
4. Don't be annoying and ask too many questions.
5. Start with naive brute force approach. First thing that comes to your mind. No need to code for this. This shows you are thingking critically and well.
6. Tell them why this approach is not better (take O(n^2) or will be not very readable or something like that.)
7. Walk through your approach and try to comment and break things. Focus on bottle necks  ( on with highest time complexity) and repeated calculations.
8. Write down the steps you are going to follow before starting to code.
9. Modularize your code from very beginning and comment.
10. Start writing your code and functions. If you don't know how to solve particular method of the proble, just wirte a dummy for it. Organize your thoughts before starting to write.
11. Don't use confusing names like i and j.
12. Think about error checks in the functions, check for all the possible input scenarios. Comment what checks you are going to write.
13. Test your code: Check for no params, 0, null, aysnc, large arrays etc..
14. Talk to the interviewer about how to improve the code, tell him what part you might google to get a better solution, ask him what is the most intresting soultion he saw for this problem.
15. If a interviewer asks for the solution to large data sets, as an extension, then tell him or rather  show him, the divide and conquer approach ofreading only chunks of input at a time, writing output back to disk and later combining them.

**Summary:**

1. Think out loud
2. Ask for clarifications
3. Think about edge cases
4. Test right there
5. Talk before writing anything

Arrays can be static or dynamic. Static arrays are fixed in size. We usually use dynamic arrays in interviews. Treat String questions as Array questions.


For reversal of a string follwing code can be used:

```
import java.util.Scanner;
public class Arrays {
  public static void main(String[] args) {
    System.out.println("Please enter a string: ");
    Scanner sc = new Scanner(System.in);
    String sent = sc.nextLine();
    sc.close();
    System.out.print(reverseString(sent));
  }
  public static String reverseString(String sent) {
    StringBuilder sb = new StringBuilder();
    for(int i = sent.length()-1; i>=0; i--)
        sb.append(sent.charAt(i));
    return sb.toString();
  }
  public static String reverseString2(String sent) {
    return new StringBuilder(sent).reverse().toString();
  }
}

```
 For merging of two sorted arrays, foolwing code can be used:



```
public class Arrays {
  public static void main(String[] args) {
    int[] arr1 = new int[]{0,3,4,31};
    int[] arr2 = new int[]{4,6,30};
    int[] merged = mergeSortedArrays(arr1,arr2);
    for(int ind = 0; ind<merged.length; ind++)
      System.out.println(merged[ind]);
  }
 public static int[] mergeSortedArrays(int[] arr1, int[] arr2) {
    if(arr1.length == 0)
        return arr2;
    else if(arr2.length == 0)
        return arr1;
    int[] merged = new int[arr1.length+arr2.length];
    int i = 0,j=0,k=0;
    while(i<arr1.length && j<arr2.length) {
      if(arr1[i]<arr2[j])
          merged[k++] = arr1[i++];
      else
          merged[k++] = arr2[j++];
    }
    while(i<arr1.length)
      merged[k++] = arr1[i++];
    while(i<arr2.length)
      merged[k++] = arr2[j++];
    return merged;
  }
}

```
